```python
import random

print("Welcome to")
print('''
 _
| |
| |__   __ _ _ __   __ _ _ __ ___   __ _ _ __
| '_ \ / _` | '_ \ / _` | '_ ` _ \ / _` | '_ \
| | | | (_| | | | | (_| | | | | | | (_| | | | |
|_| |_|\__,_|_| |_|\__, |_| |_| |_|\__,_|_| |_|
                    __/ |
                   |___/  ''','''
   _______
     |/   |
     |   (_)
     |   \|/
     |    |
     |   / \
     |
 jgs_|___''')

# hangmna stages (6 lives)
stages = ['''
 +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========''','''
 +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========''', '''
 +---+
  |   |
  O   |
 /|\  |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========''', '''
 +---+
  |   |
  O   |
      |
      |
      |
=========''', '''
+---+
  |   |
      |
      |
      |
      |
=========''']

# word list
word_list =["sejal","shreyash","bubu","pasta","momo","frenchfries","momos","cheesedosa","panipuri","biryani","stella","rajroyal","love","maharajbagh","trilliummall","icecream"]

#choose random word
chosen_word = random.choice(word_list)


# put - for chosen_word
placeholder = ""
word_length = len(chosen_word)
for position in range(word_length):
  placeholder += "-"
print(placeholder)


# 3 to continue many times
# game loop
game_over = False
correct_letter = [] # correct words will be added to this
lives = 6


while not game_over:
  guess = input("Guess a letter: ")

  if guess in correct_letter:
    print(f"You have already guessed {guess}")
  display = ""
  # 1 guessed right or wrong , 2 display letter, 3 keep the correct letters
  for letter in chosen_word:
    if letter == guess:
      display += letter
      correct_letter.append(letter)

    elif letter in correct_letter:
      display += letter

    else:
      display += "-"
  print(display)

  if guess not in chosen_word:
      lives -= 1
      print(f"You guessed {guess},that's not in word. You lose a life.")
      if lives == 0:
        game_over = True
        print("**************You lose**************")

  if "-" not in display:
    game_over = True
    print("**************You win**************")

  print(stages[lives])



```
